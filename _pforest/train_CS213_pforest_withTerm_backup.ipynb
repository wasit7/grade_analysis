```python
import os
sname='CS213' #subject_name
if not os.path.exists(sname):
    os.makedirs(sname)
```

```python
import pandas as pd

df= pd.read_pickle("../df_m20/df5.pkl")
df=df[df['COURSEID']==sname]
#df["COURSEID"].value_counts()
#df=pd.read_csv('../df_m20/df_CS213.csv',index_col=0)
df
```

```python
x_head=df.columns[3:]
y_head=['GRADE']
print x_head
print y_head
x=df.as_matrix(x_head)
y=df.as_matrix(y_head).T[0]
```

```python
import pickle
from sklearn.cross_validation import StratifiedKFold
sfk=StratifiedKFold(y, n_folds=5, shuffle=True)
for i,(train_index, test_index) in enumerate(sfk):
    print "i:%s, train_index:%s, test_index:%s"%(i,train_index,test_index)
    #dataset00.pic
    with open(sname+'/dataset%02d.pic'%(i), 'wb') as pickleFile:
        theta_dim=1
        I=x[test_index]
        L=y[test_index]
        ###please check Clmax
        clmax=14
        theta_range=I.shape[1]
        pickle.dump((clmax,theta_dim,theta_range,len(L),L,I,None), pickleFile, pickle.HIGHEST_PROTOCOL)
```

```python
from pforest.master import master
m=master(dsetname=sname)
m.reset()
m.train()
with open('%s/out_tree1.pic'%(sname), 'wb') as pickleFile:
    pickle.dump(m.root, pickleFile, pickle.HIGHEST_PROTOCOL)
```

```python
def fold_index(k=5,N=10):
    x=np.random.permutation(np.arange(N))
    return np.split(x, [(i)*N/k for i in xrange(1,k)])
fold_index(5,765)
```

```python
import pickle
from pforest.master import master
import numpy as np
from pforest.tree import tree

n_folds=5
N=len(y)
clmax=14
theta_dim=1
sfk=fold_index(k=n_folds,N=N)
recall=np.zeros(n_folds)
for k in xrange(n_folds):
    #prepare the dataset for every core and evaluation
    for j in xrange(n_folds):
        ix=(k+j)%5
        sample_index=sfk[ix]
        with open(sname+'/dataset%02d.pic'%(ix), 'wb') as pickleFile:
            I=x[sample_index]
            L=y[sample_index]
            theta_range=I.shape[1]
            pickle.dump((clmax,theta_dim,theta_range,len(L),L,I,None), pickleFile, pickle.HIGHEST_PROTOCOL)
            
    #train
    m=master(dsetname=sname)
    m.reset()
    m.train()
    with open(sname+'/out_tree%02d.pic'%(k), 'wb') as pickleFile:
        pickle.dump(m.root, pickleFile, pickle.HIGHEST_PROTOCOL)
            
    #load last dataset
    with open(sname+'/dataset%02d.pic'%(n_folds-1), 'rb') as pickleFile:
        d=pickle.load(pickleFile)
        L_last=d[4]
        I_last=d[5]
    ##evaluate    
    actual=np.array(L_last)
    predict=np.zeros(len(L_last))
    t=tree()
    t.settree(m.root)
    for i in xrange(len(L_last)):
        Prob=t.classify(I_last[i])
        predict[i]=np.argmax(Prob)

        #print "actual grade:%d predicted grade:%d"%(actual[i],predict[i])
    ##confustion matrix
    cm = np.zeros((clmax,clmax))
    for i in xrange(len(L)):
        cm[predict[i],actual[i]] +=1
    print cm
    tp=0
    for i in xrange(clmax):
        tp += cm[i,i]
    recall[k]=tp/cm.sum().sum()
    print "fold: %d, recall:%.3f"%(k, recall[k])
```
